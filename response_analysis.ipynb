```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zoom_path = "data/out/zoom.csv"
f2f_path = "data/out/f2f.csv"
```

```python
zoom_df = pd.read_csv(zoom_path)
zoom_df
```

```python
f2f_df = pd.read_csv(f2f_path)
f2f_df
```

```python
q1_df = f2f_df.iloc[:, 7: 27]
q1_df
```

```python
q1_df.isnull().values.any()
```

```python
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity

kmo_all, kmo_model = calculate_kmo(q1_df)
bartlett_sphericity_chi_square, bartlett_p_value = calculate_bartlett_sphericity(q1_df)

print(f"KMO Test: {kmo_model}")
print(f"Bartlett's Test: chi-square={bartlett_sphericity_chi_square}, p-value={bartlett_p_value}")
```

```python
from factor_analyzer import FactorAnalyzer

# Create a factor analysis object and perform factor analysis
fa = FactorAnalyzer(rotation=None, n_factors=q1_df.shape[1], method='minres')
fa.fit(q1_df)

# Check Eigenvalues
ev, _ = fa.get_eigenvalues()
print(ev)

# Counting values greater than 1
count_greater_than_one = np.sum(ev > 1)

print(count_greater_than_one)
```

```python
# Assuming you've chosen 5 factors based on eigenvalues
fa = FactorAnalyzer(rotation='varimax', n_factors=2, method='minres')
fa.fit(q1_df)

# Get the loadings (the correlation coefficient for each question with each factor)
loadings = fa.loadings_
print(loadings)

# Get variance of each factor
variance = fa.get_factor_variance()
print(variance)
```

```python
# Creating the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(loadings, annot=True, cmap='viridis', fmt=".2f",
            cbar=True, xticklabels=[f'Factor {i+1}' for i in range(7)],
            yticklabels=[f'Item {i+1}' for i in range(20)])
plt.title('Heatmap of Factor Loadings')
plt.xlabel('Factors')
plt.ylabel('Questionnaire Items')
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

n_components = 5

# Applying PCA
pca = PCA(n_components=n_components)  # Choosing 7 components as an example
pca.fit(q1_df.values)

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_

# Cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance)

# Plotting the explained variance
plt.figure(figsize=(24, 6))

plt.subplot(1, 2, 1)
plt.bar(range(1, n_components + 1), explained_variance, alpha=0.5, align='center',
        label='Individual explained variance')
plt.step(range(1, n_components + 1), cumulative_explained_variance, where='mid',
         label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component')
plt.legend(loc='best')
plt.title('Explained Variance by Component')

# Heatmap of the loadings (component vectors)
loadings = pca.components_

plt.subplot(1, 2, 2)
sns.heatmap(loadings, cmap='coolwarm', annot=True, fmt=".2f",
            xticklabels=[f'Question {i+1}' for i in range(0, 20)],
            yticklabels=[f'Component {i+1}' for i in range(0, n_components)])
plt.title('Heatmap of PCA Component Loadings')

plt.tight_layout()
plt.show()
```

```python
# Projecting the standardized data onto the first two principal components
pca_scores = pca.transform(q1_df.values)

# Extracting the scores for the first two components
component_1_scores = pca_scores[:, 0]  # First component scores
component_2_scores = pca_scores[:, 1]  # Second component scores

# Plotting the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(component_1_scores, component_2_scores)
plt.title('PCA: First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.show()
```
